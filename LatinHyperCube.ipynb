```python
f = open('../../LosAlamos/PRISM-1.2.0/prism-1.2/input/data/s1n_frdm2012.bin', 'rb')
#f = open('test.bin', 'rb')

f.seek(4)
recl = np.fromfile(f, dtype=np.dtype('i4'), count=1)
print(recl)

data_frdm = np.zeros((9189,3))
dict_frdm = {}

for i in range(9189):
    f.seek(16+72*i)
    rZ = np.fromfile(f, dtype=np.dtype('i4'), count=1)

    f.seek(28+72*i)
    rN = np.fromfile(f, dtype=np.dtype('i4'), count=1)

    f.seek(40+72*i)
    pZ = np.fromfile(f, dtype=np.dtype('i4'), count=1)

    f.seek(56+72*i)
    pN = np.fromfile(f, dtype=np.dtype('i4'), count=1)

    f.seek(72+72*i)
    s1n = np.fromfile(f, dtype=np.dtype('f8'), count=1)
    
    data_frdm[i] = (rZ[0],rN[0],s1n[0])
    dict_frdm[(rZ[0],rN[0])] = s1n[0]

f.close()
```

```python
data_frdm_var = np.zeros(354)
dict_frdm_var = {}
dict_frdm_label = {}

for i in range(0,354,1):
    dict_frdm_var[i]={}

i = 0
for A in range(150,180,1):
    for N in range(95,115,1):
        Z = A - N
        pos = [N,Z]
        if ((N<=94) or (N>=95 and N<= 98 and Z < 66) or (N>=99 and N<= 100 and Z < 68) or (N>=101 and N<= 104 and Z < 70) or (N>=105 and N<=108 and Z < 72) or (N>108)) and (dict_frdm.get((Z,N)) != None) and (dict_ame.get((Z,N)) == None):
            dict_frdm_var[i] = dict_frdm.get((Z,N))
            data_frdm_var[i] = dict_frdm.get((Z,N))
            dict_frdm_label[i] = [Z,N]
            i = i+1
```

```python
from pyDOE import *
import numpy as np
np.random.seed(100)
lhd = lhs(len(data_frdm_var), samples=20000, criterion='center')
print(lhd[:,0])
```

```python
var_size = 1 #MeV
for i in range(len(data_frdm_var)):
    lhd[:,i] = (lhd[:,i]*var_size-var_size/2)+data_frdm_var[i]
```

```python
#ascii
for i in range(len(lhd[:,0])):
    f = open('varied/s1n_frdm2012var%d.dat' % i,'w')
    f.write("%d\n" %len(data_frdm_var))

    for j in range(len(data_frdm_var)):
        rZ = dict_frdm_label[j][0]
        rN = dict_frdm_label[j][1]
        pZ = rZ
        pN = rN-1
        nZ = 0
        nN = 1
        s1n = lhd[i,j]

    
        # Finally write to file here
        f.write("%3d\n"%(rZ))
        f.write("%3d\n"%(rN))
        f.write("%3d %3d\n"%(pZ,nZ))
        f.write("%3d %3d\n"%(pN,nN))
        f.write("%f\n"%(s1n))    
        #f.write("%f\n"%(rZ[0]+rN[0]))
        #f.write("%f\n"%(p[0]*(pZ+pN[0])+p[1]*(pZ+pN[1])+p[2]*(pZ+pN[2])+p[3]*(pZ+pN[3])+num_neutron))
    f.close()
```

```python
#bin 
import numpy as np
from scipy.io import FortranFile
for i in range(len(lhd[:,0])):
    f = FortranFile('varied/s1n_frdm2012var%d.bin' % i,'w')
    f.write_record(np.array(len(data_frdm_var), dtype=np.dtype('i4')))
    
    for j in range(len(data_frdm_var)):
        rZ = dict_frdm_label[j][0]
        rN = dict_frdm_label[j][1]
        pZ = rZ
        pN = rN-1
        s1n = lhd[i,j]
        
        f.write_record(np.array(rZ, dtype=np.dtype('i4')))  # Reactant Z's
        f.write_record(np.array(rN, dtype=np.dtype('i4')))  # Reactant N's
        f.write_record(np.array(pZ, dtype=np.dtype('i4'))) # Product Z's
        f.write_record(np.array(pN, dtype=np.dtype('i4'))) # Product N's
        f.write_record(np.array(s1n, dtype=np.dtype('f8')))  # Property (S1n)
        
    f.close()
```
